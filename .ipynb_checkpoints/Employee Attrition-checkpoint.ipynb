# Employee Attrition Classification 

## Abstract

Employee attrition can have ruinous impact on the health of a company. 
Whenever an experience employee leaves the organization, there is a cost associated with it. 
A part of this cost is quantifiable numbers and the other part is hidden costs. The quantifiable cost includes the cost incurred in recruiting a replacement and training them. It would be great to have a machine learning model that will help the management to identify the employees who may potentially leave or quit. So that the management can take corrective measures to avoid employee attrition. The intend of the project is to use different classification models and tools available to accurately predict whether an employee will leave the company or not. The various classification techniques used in this project are Logistic Regression classifier, Support vector Machine classifier, Neural Networks and Ensemble methods (Decision Forests). The tools used are python in Jupyter note book and Microsoft Azure Machine Learning 



## Introduction

High employee attrition rate is a bad news for any organization. **Employee turnover rate** is number of employees who left during the entire year divided by the total number of employees at the beginning of the year and it is calculated annually. A high employee turnover may be risky for health of the company as high cost are associate with it. The quantifiable cost includes the cost involves recruiting, training etc. Generally, the productivity of the new hire is lower until he/she learn the business. For a customer facing role, if some customer/s are unhappy with the new hire,
the business could lose customers who are dissatisfied with service leading to a loss in the revenue.
Thus, it is very crucial for company’s management to understand the factors leading to employee attrition 
and to identify the potential employees who could quit and finally to institutionalized measures to 
reduce attrition rates.

There can be umpteen reasons for employee attrition. Many of these reasons are personal and may varies from person to person, it can be as diverse as different human personalities. But it is important to analyze some of the basic factors that triggers most resignations?

As discussed, the primary goal of this project is to correctly identify the potential employees in a company who could leave using a classification model.




### Reasons for Employee Attrition

To take corrective measure to stop high attrition rate in an organization, it is important to analyze what are the factors that influences a person’s decision to quite the company. Here are some of the reasons, why an employee might leave his/her current job :-

    
1.	Mostly people leave their employer for an increased paycheck or career advancement,
2.	Many researches also shows that often the search for a new position is trigger due to dissatisfaction with an immediate manager.
3.	Employee are not engaged and they are bored. High-performing workers need to feel that they are being challenged and are moving forward in terms of professional growth and development.

###  Data Exploration and Visualization of Dataset

I have created the two functions EA_snsbox, EA_bar for creating the boxplots and bar charts on various predictors variables

```python
# Creates boxplots on the data
def EA_snsbox(df,col):
    import matplotlib
    import numpy as np
    import seaborn as sns
    import matplotlib.pyplot as plt
    sns.set_style("dark")
    g=sns.boxplot(x="Attrition",y=col,data= df,palette="Set1")
    sns.plt.title('Box plot of ' + col + '\n for Employee Attrition')
    g.set_xticklabels(labels=df["Attrition"], rotation=90)
    
# creates the bar charts on the data
def EA_bar(df, cols):
    import seaborn as sns
    import matplotlib.pyplot as plt
    pf=df.groupby([cols,"Attrition"]).size().reset_index(name='counts')
    sns.set_style("dark")
    g=sns.barplot(x=cols, y="counts",hue="Attrition", data=pf)
    sns.plt.title(cols +' vs. \n Employee Attrition')
    sns.plt.ylabel('Count')
    g.set_xticklabels(labels=np.unique(pf[cols]), rotation=90)
    
# Interaction between various variables
def EA_VP(df,col1,col2):
    import seaborn as sns
    import matplotlib.pyplot as plt
    sns.set_style("dark")
    ax = sns.violinplot(x=col1, y=col2, hue="Attrition",
                        data=df, palette="Set2", split=True,scale="count")
        
```

### Pre-Processing of Dataset

It is important to pre-process the data before doing the actually analysis. This steps involves :-
    1. Handling the missing data
    2. Droping the redundant columns.
    3. Scaling or Normalization of data (as needed)

**1.Droping the redundant columns:**
    
* All the employees of company are over 18 years of age.  
* Standard Hours is constant value of 80  
* 'Employee Count' is always 1   
* 'Employee Number' - unique identification of employee.  
* The variables 'DailyRate','EducationField','HourlyRate', 'JobRole','MonthlyRate','YearsAtCompany','PerformanceRating' are columns of zero importance of our datasets (verified using Permutation Feature Importance Module IN Azure ML )

```python
import pandas as pd
import numpy as np

Emp_Attrition= Emp_Attrition.drop(['Over18','StandardHours',
'EmployeeCount','EmployeeNumber','DailyRate','EducationField','HourlyRate',
'JobRole','MonthlyRate','YearsAtCompany','PerformanceRating' ],axis=1)
# Independent Variable
X = Emp_Attrition
X= X.drop(X.columns[[1]],axis=1)
# dependent Variable
Y= Emp_Attrition.iloc[:, 1]
```

**2.Coverting categorical data into number:**

Many Machine Learning tools accept only numbers as input, this can cause a problem as the above dataset has some categorical predictor variables too. To convert them as number I have used the 'LabelEncoder'. To convert the nominal variables like ‘Marital status’, ‘Gender’,’OverTime’ etc. I used the 'One-Hot-encoder' form sklearn.preprocessing.



```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
# Converting categorical values into numerical values
labelencoder_x= LabelEncoder()
for i in [1,2,6,10,13]:
    X.iloc[:, i]=labelencoder_x.fit_transform( X.iloc[:, i])
    
# creating columns for categorical values
onehotencoder =OneHotEncoder(categorical_features=[1,6,10,13])
X=onehotencoder.fit_transform(X).toarray() 
# yes/no into numerical values
Y=labelencoder_x.fit_transform(Y)

    
```

**3. Spliting Data into train and Test**

To get a honest assessments of the performance of our predictive model, it is essential to split the data into a train and
test dataset.


```python
# spliting the data into test and training set
from sklearn.model_selection import  train_test_split
# 20% of test data
X_train,X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,random_state=12345)

```

# CLASSIFICATION #


### 1. RANDOM FOREST

Random forest classifier is an ensemble learning method for classification. The number of tree in the forest 
(n_estimators) =100. The 'Gini impurity' is used to measure the quality of split.
                                                                                                        

```python
from sklearn.ensemble import  RandomForestClassifier
rf_model= RandomForestClassifier(n_estimators=100, 
                                 oob_score=True,
                                 criterion= 'gini',
                                 max_features=7,
                                 random_state=16333336)

rf_model.fit(X_train, Y_train)
print(rf_model.feature_importances_)
```

**Cross Validation with K=10 folds**

```python
import sklearn.model_selection as  model_selection
kfold = model_selection.KFold(n_splits =10,random_state=1452345)

#cross validation
scores = model_selection.cross_val_score(rf_model, X_train, Y_train, cv=kfold, scoring='accuracy')
print("Train accuracy %0.2f (+/- %0.2f)" % (scores.mean()*100, scores.std()*100))

# Predicting the test output
from sklearn.metrics import accuracy_score,roc_curve, auc
preds = model_selection.cross_val_predict(rf_model, X_test, Y_test, cv=kfold)
print("Test accuracy %0.2f" % (100* accuracy_score(Y_test, preds)))
```

**Area under the Curve(AUC)**:

```python
rf_model= RandomForestClassifier(n_estimators=100, 
                                 oob_score=True,criterion= 'gini',
                                 max_features=7,random_state=35666666)

rf_model.fit(X_train,Y_train)
fpr_rf, tpr_rf, _ = roc_curve(Y_test, rf_model.predict_proba(X_test)[:,1])
# Calculate the AUC
roc_auc_r = auc(fpr_rf, tpr_rf)
print('ROC AUC: %0.2f' % roc_auc_r)
```

### 2. Logistic Regression 

Before apply the Logistic Regression classifier and support vector Machine classifier, first we need to scale or normalized the data.

Logistic regression also known as **logit regression** is a regression model where the dependent variable is categorical variable. Here 'C' is inverse of regularization strength, smaller values specify stronger regularization. The function for tuning the parameters available in scikit-learn is called gridSearchCV() is used to get bets parameters.

**param_grid**: It is the dictionary or list with parameters names (string) as keys and lists of parameter settings to try as values. The penalty used is "l2"



```python
# scaling data
from sklearn.preprocessing import StandardScaler
sc_X= StandardScaler()
X_train =sc_X.fit_transform(X_train)
X_test =sc_X.transform(X_test)
```

**Tuning of Parameters**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid_l={'C':[0.001,0.01,0.1,1,10,100,1000],
              'tol':[0.1,0.01,0.001,0.0001]}
CV_log = GridSearchCV(cv=10,
                      estimator=LogisticRegression(C=1.0,penalty='l2',tol=0.001)
                      , param_grid= param_grid_l)

CV_log.fit(X_train,Y_train)
print(CV_log.best_params_)
```

**Cross Validation with K=10 folds**

```python
from sklearn import model_selection
scores =model_selection.cross_val_score(CV_log, X_train, Y_train, cv=kfold, scoring='accuracy')

print("Train accuracy %0.2f (+/- %0.2f)" % (scores.mean()*100, scores.std()*100))

# Predicting the 
from sklearn.metrics import accuracy_score,roc_curve, auc
cv_preds=list()
preds = model_selection.cross_val_predict(CV_log, X_test, Y_test, cv=kfold)

print("Test accuracy %0.2f" % (100* accuracy_score(Y_test, preds)))
```

**Area under the Curve(AUC):**

```python
logR_clf= LogisticRegression(C=1000,penalty='l2',tol=0.1)
logR_clf.fit(X_train,Y_train)
fpr_l, tpr_l, _ = roc_curve(Y_test, logR_clf.predict_proba(X_test)[:,1])
# Calculate the AUC
roc_auc_l = auc(fpr_l, tpr_l)
print('ROC AUC: %0.2f' % roc_auc_lg)
```

### 3. Support Vector Machine Classifer

The class used for SVM classification in scikit-learn is svm.SVC(). The parameter for SVM classifier:    
1. C: It is the regularization parameter, C, of the error term.
2. kernel: It specifies the kernel type to be used in the algorithm. It can be ‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’, or a callable. 
    The default value is ‘rbf'.
3. degree: It is the degree of the polynomial kernel function (‘poly’) and is ignored by all other kernels.
    The default value is 3.
4. gamma: It is the kernel coefficient for ‘rbf’, ‘poly’, and ‘sigmoid’. 
    If gamma is ‘auto’, then 1/n_features will be used instead.
    
    
The SVM is tune by changing parameters C, gamma and the kernel function.
The function for tuning the parameters is called gridSearchCV().
param_grid: It is the dictionary or list with parameters names (string) as keys and lists of parameter settings to try as values.


**Tunning of parameter**

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
svc=SVC()
param_grid= {
        "kernel":['linear','rbf'],
        "gamma":[0.01,0.02,0.05,0.1],
        "C" : [1,10,100]}
CV_svc = GridSearchCV(estimator=svc, param_grid= param_grid,cv=10)
CV_svc.fit(X_train,Y_train)
print(CV_svc.best_params_)
cv_scores = list()
from sklearn import model_selection
svc2= SVC()
svc=SVC(C =10 ,kernel ="rbf",gamma=0.1)
kfold = model_selection.KFold(n_splits =10,random_state=16792345)
print("done")
```

```python
from sklearn import model_selection
scores =model_selection.cross_val_score(svc, X_train, Y_train, cv=kfold, scoring='accuracy')

cv_scores.append(scores.mean()*100)
print("Train accuracy %0.2f (+/- %0.2f)" % (scores.mean()*100, scores.std()*100))

# Predicting the 
from sklearn.metrics import accuracy_score,roc_curve, auc
cv_preds=list()
preds = model_selection.cross_val_predict(svc, X_test, Y_test, cv=kfold)
cv_preds.append(accuracy_score(Y_test, preds)*100)
print("Test accuracy %0.2f" % (100* accuracy_score(Y_test, preds)))
```

**Area under the Curve(AUC):**

```python
import matplotlib.pyplot as plt
svc=SVC(C =10 ,kernel ="rbf",gamma=0.01,probability= True)
svc.fit(X_train,Y_train)
from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(Y_test, svc.predict_proba(X_test)[:,1])
# Calculate the AUC
roc_auc_s = auc(fpr, tpr)
print('ROC AUC: %0.2f' % roc_auc_s)
```

### Comparing Support Vector Machine, Logistic Regression and Random Forest Classifier

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkgreen',label='SVM-classifier (area = %0.2f)' % roc_auc_s)
plt.plot(fpr_rf, tpr_rf, color='darkorange',label='RandomForest(area = %0.2f)' % roc_auc_r)
plt.plot(fpr_l, tpr_l, color='darkblue',label='Logistic Regression(area = %0.2f)' % roc_auc_l)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
